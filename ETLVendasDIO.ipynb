#Extract

Inicialmente, é feita a importação das bibliotecas necessárias, incluindo sqlite3 para interagir com o banco de dados SQLite, pathlib para trabalhar com caminhos de arquivo e pandas para manipulação de dados

```python
import sqlite3
from pathlib import Path
import pandas as pd
```

 A seguir, o arquivo em formato csv e chamado 'Vendas.csv' é lido usando a biblioteca já importada pandas, ou seja, o arquivo é extraído. Esse arquivo contém informações sobre vendas de produtos, tais como: do produto, quantidade e preço unitário.

```python

vendas = pd.read_csv('Vendas.csv', sep=';', usecols=['Produtos','Quantidade','Preco_Unitario'])
vendas
```

#Transform

Nessa etapa, é realizada as seguintes tranformações: Na coluna 'Preco_Unitario', o símbolo R$ é removido, as vírgulas são substituídas por pontos e os valores da coluna são convertidos para float. Além disso, uma nova coluna chamada 'Total' é criada, multiplicando as colunas 'Quantidade' e 'Preco_Unitario'.

```python
vendas['Preco_Unitario'] = vendas['Preco_Unitario'].str.replace('R\$','').str.replace(',','.')
vendas['Preco_Unitario'] = vendas['Preco_Unitario'].astype(float)
vendas
```

Aqui, uma nova coluna chamada 'Total' é criada, multiplicando as colunas 'Quantidade' e 'Preco_Unitario'.

```python
vendas['Total'] = vendas['Quantidade']* vendas['Preco_Unitario']
vendas
```

#Load

Na última etapa, que é o carregamento, um arquivo de banco de dados SQLite chamado 'vendas.db' e os dados transformados são carregados no banco sqlite, resultando em uma tabela 'vendas' no banco de dados, são consultados e exibidos usando SQL.


```python
Path("vendas.db").touch()

db_path = "vendas.db"
conn = sqlite3.connect(db_path)
cursor = conn.cursor()
```

```python
vendas.to_sql(name = 'vendas', con = conn, if_exists = 'replace')
```

```python
query = """

select
*
from vendas ;
"""

df = pd.read_sql(query, conn)
display(df)
```
